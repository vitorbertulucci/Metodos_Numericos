```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def f(x):
    return -x * np.e ** -x + 0.2

def flinha(x):
    return np.e ** (-x) *(x -1)

x = np.linspace(0,10,100)
y = f(x)

def p(A, B):
    plt.plot(x, y)
    plt.grid()
    [xmin, xmax, ymin, ymax] = plt.axis()
    plt.plot([A,A], [ymin, ymax], 'k-')
    plt.plot([B,B], [ymin, ymax], 'k-')

# Definir intervalos iniciais
A = 1
B = 10
x_tol = 0.00001
y_tol = 0.00001
x_anterior = B

def biss(A, B, x_anterior):
    # encontrar a prxima aproximao da raiz
    xi = (A + B) / 2.0 # bisseco
#     xi = (A * f(B) - B * f(A)) / (f(B) - f(A)) # posio falsa

    # visualizar o grfico
    p(A, B) 
    plt.plot([xi, xi], [ymin, ymax], 'k--')
    plt.show()
    
    condition = f(A) * f(xi)
    
    if abs(f(xi)) < y_tol:
        return f(xi)
    if abs(x_anterior - xi) < x_tol:
        return f(xi)
    
    x_anterior = xi
    
    if condition < 0:
        return biss(A, xi, x_anterior)
    elif condition > 0:
        return biss(xi, B, x_anterior)
    else:
        return xi

print("ta aqui %.6lf" % biss(A, B, x_anterior))

xi = 2

x_tol = 0.0001
y_tol = 0.0001
x_anterior = xi
while True:
    p(0,0)
    [xmin, xmax, ymin, ymax] = plt.axis()
    plt.plot([xi, xi], [ymin, ymax], 'k--')
    plt.show()
    print('Aproximao: %.6lf' % f(xi))
    
    xi = xi - f(xi) / flinha(xi)
    if abs(f(xi)) < y_tol:
        break
    if abs(x_anterior - xi) < x_tol:
        break
    
# Termo de Recurso xi
# for i in range(5):
#     xi = (A + B) / 2.0
    
#     p(A, B)
#     plt.plot([xi, xi], [ymin, ymax], 'k--')
    
#     if f(A) * f(B) < 0:
#         B = xi
#     elif f(A) * f(B) == 0:
#         print('Raiz encontrada: %.6lf' % xi)
#     else:
#         A = xi
#     plt.show()
```
