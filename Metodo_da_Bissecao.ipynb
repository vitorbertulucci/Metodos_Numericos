```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def f(x):
    return -x * np.e ** -x + 0.2

x = np.linspace(0,10,100)
y = f(x)

def p(A, B):
    plt.plot(x, y)
    plt.grid()
    [xmin, xmax, ymin, ymax] = plt.axis()
    plt.plot([A,A], [ymin, ymax], 'k-')
    plt.plot([B,B], [ymin, ymax], 'k-')

# Definir intervalos iniciais
A = 1
B = 10

# Termo de Recursão xi
for i in range(5):
    p(A, B)
    xi = (A + B) / 2.0
    plt.plot([xi, xi], [ymin, ymax], 'k--')
    if f(A) * f(B) < 0:
        B = xi
    elif f(A) * f(B) == 0:
        print('Raiz encontrada: %.6lf' % xi)
    else:
        A = xi
    plt.show()
```
